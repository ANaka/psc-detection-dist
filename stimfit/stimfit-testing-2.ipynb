```python
%matplotlib qt
s = 80
end_ind = 1999;
deconv_bounds = 2.8
fig = plt.figure(figsize=(4,25), dpi=72)
times_hist_ours = np.zeros((mat_data['traces'].shape[0],41))
times_hist_deconv = np.zeros((mat_data['traces'].shape[0],41))
all_times_deconv = np.empty((0))
all_times_cb = np.empty((0))
all_times_ours = np.empty((0))
times_hist_cb = np.zeros((mat_data['traces'].shape[0],41))
for i in range(mat_data['traces'].shape[0]):
    offset = (i-1)*115
    deconv = stfio.detect_events(mat_data['traces'][i,],template,.05,mode="deconvolution")
    crit = stfio.detect_events(mat_data['traces'][i,],template,.05,mode="criterion")
    event_times_deconv = stfio.peak_detection(deconv,deconv_bounds,1)
    event_times_crit = stfio.peak_detection(crit,3,1)
    

    these_indices = event_times_deconv[np.where(np.logical_and(event_times_deconv >= 7560, event_times_deconv <= 7580))] - 7560
    times_hist_deconv[i,these_indices] = times_hist_deconv[i,these_indices] + 1
    all_times_deconv = np.append(all_times_deconv, these_indices)
    
    these_indices = event_times_crit[np.where(np.logical_and(event_times_crit >= 7560, event_times_crit <= 7580))] - 7560
    times_hist_cb[i,these_indices] = times_hist_cb[i,these_indices] + 1
    all_times_cb = np.append(all_times_cb, these_indices)
    
    times_hist_ours = posterior_ar2[i,1530:1570]
    local_dist = posterior_ar2[i,1560:1580]/np.sum(posterior_ar2[i,1560:1580])
    #all_times_ours = np.append(all_times_ours,np.dot(range(local_dist.shape[0]),local_dist))
    all_times_ours = np.append(all_times_ours,np.argmax(posterior_ar2[i,1560:1580]))
        
    cb = plt.scatter(event_times_crit,50*np.ones(event_times_crit.shape)-offset,c = 'green',s=s/2.,edgecolor='',label='Deconvolution')
    dec = plt.scatter(event_times_deconv,25*np.ones(event_times_deconv.shape)-offset,c = 'red',s=s/2.,edgecolor='',label='Clement Bekkers')
    ours = plt.scatter(range(6000,posterior_ar2.shape[1]+6000),0*np.ones(posterior_ar2[i,].shape)-offset,c = 'blue',s=posterior_ar2[i,]/np.max(posterior_ar2[i,])*s,edgecolor='',label='Bayesian Posterior')
    #true = plt.scatter(mat_data['true_event_times'][i][0],-10*np.ones(mat_data['true_event_times'][i][0].shape)-offset,c = 'black',s=s,edgecolor='',label = 'True Times')
    obs = plt.plot(mat_data['traces'][i,]-20-offset-np.median(mat_data['traces'][i,]),c='black',label='Observation')
    
plt.xlim((6700,8500))
plt.ylim((-offset-200,200))
           
           
           
           
           
           
           
plt.legend((cb, dec, ours),('Clements Bekkers','Deconvolution','Bayesian Posterior','True Times'),loc='best',bbox_to_anchor=(1.12, 1.13),scatterpoints=1)
plt.axis('off')
plt.savefig('/media/shababo/Layover/projects/mapping/code/psc-detection/data/figures/paired-patch-all-meths.eps',format='eps')
```

```python
min_spacing = np.array([1, 10, 50, 100])
threshold = np.array([2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6])

event_times = np.empty((threshold.size,min_spacing.size),dtype='object')

traces = mat_data['traces']
deconv = np.zeros(traces.shape)
for (i,trace) in enumerate(traces):
    deconv[i,] = stfio.detect_events(trace,template,.05,mode="deconvolution")

for (j, thresh) in enumerate(threshold):
    for (k, min_space) in enumerate(min_spacing):
        event_times[j,k] = np.empty((traces.shape[0]),dtype='object')
        for (i, trace) in enumerate(traces):
            event_times[j,k][i] = stfio.peak_detection(deconv[i,], thresh, min_space)
```

```python
savefile = '/media/shababo/Layover/projects/mapping/code/psc-detection/stimfit/stimfit-data/deconv-detection-results-rev.mat'
scio.savemat(savefile, dict(event_times=event_times, threshold=threshold,min_spacing=min_spacing))
```

```python
min_spacing = np.array([1, 10, 50, 100])
threshold = np.array([1., 1.5, 2.0, 2.5, 3, 3.5])

event_times = np.empty((threshold.size,min_spacing.size),dtype='object')

traces = mat_data['traces']
deconv = np.zeros((traces.shape[0],traces.shape[1]-template.size))
for (i,trace) in enumerate(traces):
    deconv[i,] = stfio.detect_events(trace,template,.05,mode="criterion")

for (j, thresh) in enumerate(threshold):
    for (k, min_space) in enumerate(min_spacing):
        event_times[j,k] = np.empty((traces.shape[0]),dtype='object')
        for (i, trace) in enumerate(traces):
            event_times[j,k][i] = stfio.peak_detection(deconv[i,], thresh, min_space)
```
